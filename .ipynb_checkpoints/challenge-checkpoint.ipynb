  ## Identifying Outliers using Standard Deviation

```python
# initial imports
import pandas as pd
import numpy as np
import random
from sqlalchemy import create_engine

```

```python
# create a connection to the database
engine = create_engine("postgresql://postgres:postgres@localhost:5432/fraud_detection")
query = "SELECT * FROM card_holder_transactions;"
transaction_df = pd.read_sql(query, engine)
transaction_df.head()
```

```python
# code a function to identify outliers based on standard deviation
def return_outliers(df):
    mean = df['amount'].mean()
    sd = df['amount'].std()
    return df[(df['amount'] < mean - 2 * sd) | (df['amount'] > mean + 2 * sd)]
```

```python
# find anomalous transactions for 3 random card holders
# viewing outlier for 3, 18 and 2 as you can see 2 does not show up which matches the results of the previous notebook
df = return_outliers(transaction_df[(transaction_df['id'] == 3)|(transaction_df['id'] == 18)|(transaction_df['id'] == 2)])
df
```

  ## Identifying Outliers Using Interquartile Range

```python
# code a function to identify outliers based on interquartile range
def return_outliers_quartile(df):
    q25, q75 = np.percentile(df['amount'], 25), np.percentile(df['amount'], 75)
    interquartile = q75 - q25
    cut_off = interquartile * 1.5
    lower, upper = q25 - cut_off, q75 + cut_off
    return df[(df['amount'] < lower) | (df['amount'] > upper)]
```

```python
# find anomalous transactions for 3 random card holders
df = return_outliers_quartile(transaction_df[(transaction_df['id'] == 3)|(transaction_df['id'] == 18)|(transaction_df['id'] == 2)])
df
```
